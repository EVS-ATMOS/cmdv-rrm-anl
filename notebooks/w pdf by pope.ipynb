## W pdfs by Pope regime

This code views the pdfs of vertical velocity generated by dask over the CPOL/Berrima dataset.

```python
from netCDF4 import Dataset
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
deep_west_file = '/home/rjackson/data/wpdfregime0_15000.0_20000.0varble.cdf'
moist_east_file = '/home/rjackson/data/wpdfregime1_15000.0_20000.0varble.cdf'
east_file = '/home/rjackson/data/wpdfregime2_15000.0_20000.0varble.cdf'
dry_east_file = '/home/rjackson/data/wpdfregime3_15000.0_20000.0varble.cdf'
shallow_west_file = '/home/rjackson/data/wpdfregime4_15000.0_20000.0varble.cdf'
```

```python
# Deep west
deep_west_cdf = Dataset(deep_west_file, 'r')
mean_dw = deep_west_cdf.variables['mean'][:]
median_dw = deep_west_cdf.variables['median'][:]
ninety_dw = deep_west_cdf.variables['ninety'][:]
ninety_five_dw = deep_west_cdf.variables['ninety_five'][:]
ninety_nine_dw = deep_west_cdf.variables['ninety_nine'][:]
z_dw = deep_west_cdf.variables['z'][:]
deep_west_cdf.close()

# Moist east
moist_east_cdf = Dataset(moist_east_file, 'r')
mean_me = moist_east_cdf.variables['mean'][:]
median_me = moist_east_cdf.variables['median'][:]
ninety_me = moist_east_cdf.variables['ninety'][:]
ninety_five_me = moist_east_cdf.variables['ninety_five'][:]
ninety_nine_me = moist_east_cdf.variables['ninety_nine'][:]
z_me = moist_east_cdf.variables['z'][:]
moist_east_cdf.close()

# East
east_cdf = Dataset(east_file, 'r')
mean_e = east_cdf.variables['mean'][:]
median_e = east_cdf.variables['median'][:]
ninety_e = east_cdf.variables['ninety'][:]
ninety_five_e = east_cdf.variables['ninety_five'][:]
ninety_nine_e = east_cdf.variables['ninety_nine'][:]
z_e = east_cdf.variables['z'][:]
east_cdf.close()

# Dry east
dry_east_cdf = Dataset(dry_east_file, 'r')
mean_de = dry_east_cdf.variables['mean'][:]
median_de = dry_east_cdf.variables['median'][:]
ninety_de = dry_east_cdf.variables['ninety'][:]
ninety_five_de = dry_east_cdf.variables['ninety_five'][:]
ninety_nine_de = dry_east_cdf.variables['ninety_nine'][:]
z_de = dry_east_cdf.variables['z'][:]
dry_east_cdf.close()

# Shallow west
shallow_west_cdf = Dataset(shallow_west_file, 'r')
mean_sw = shallow_west_cdf.variables['mean'][:]
median_sw = shallow_west_cdf.variables['median'][:]
ninety_sw = shallow_west_cdf.variables['ninety'][:]
ninety_five_sw = shallow_west_cdf.variables['ninety_five'][:]
ninety_nine_sw = shallow_west_cdf.variables['ninety_nine'][:]
z_sw = shallow_west_cdf.variables['z'][:]
shallow_west_cdf.close()
```

```python
plt.figure(figsize=(14,14))
plt.subplot(321)
plt.plot(mean_dw,z_dw,linewidth=2)
plt.plot(median_dw,z_dw,linewidth=2)
plt.plot(ninety_dw,z_dw,linewidth=2)
plt.plot(ninety_five_dw, z_dw,linewidth=2)
plt.plot(ninety_nine_dw, z_dw,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Deep west')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')
plt.legend(['Mean', 'Median', '90', '95', '99'])

plt.subplot(322)
plt.plot(mean_me,z_me,linewidth=2)
plt.plot(median_me,z_me,linewidth=2)
plt.plot(ninety_me,z_me,linewidth=2)
plt.plot(ninety_five_me, z_me,linewidth=2)
plt.plot(ninety_nine_me, z_me,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Moist east')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')

plt.subplot(323)
plt.plot(mean_e,z_e,linewidth=2)
plt.plot(median_e,z_e,linewidth=2)
plt.plot(ninety_e,z_e,linewidth=2)
plt.plot(ninety_five_e, z_e,linewidth=2)
plt.plot(ninety_nine_e, z_e,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('East')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')

plt.subplot(324)
plt.plot(mean_de,z_de,linewidth=2)
plt.plot(median_de,z_de,linewidth=2)
plt.plot(ninety_de,z_de,linewidth=2)
plt.plot(ninety_five_de, z_de,linewidth=2)
plt.plot(ninety_nine_de, z_de,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Dry East')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')

plt.subplot(325)
plt.plot(mean_sw,z_sw,linewidth=2)
plt.plot(median_sw,z_sw,linewidth=2)
plt.plot(ninety_sw,z_sw,linewidth=2)
plt.plot(ninety_five_sw, z_sw,linewidth=2)
plt.plot(ninety_nine_sw, z_sw,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Shallow West')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')
```
