## W pdfs by Drosdowsky (1997) classification

This code views the pdfs of vertical velocity generated by dask over the CPOL/Berrima dataset.

```python
from netCDF4 import Dataset
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
break_file = '/home/rjackson/data/wpdfregime_dros0_varble0.cdf'
monsoon_file = '/home/rjackson/data/wpdfregime_dros1_varble0.cdf'
break_file_sea = '/home/rjackson/data/wpdfregime_dros0_varble1.cdf'
monsoon_file_sea = '/home/rjackson/data/wpdfregime_dros1_varble1.cdf'
```

```python
# Break
break_cdf = Dataset(break_file, 'r')
mean_b = break_cdf.variables['mean'][:]
median_b = break_cdf.variables['median'][:]
ninety_b = break_cdf.variables['ninety'][:]
ninety_five_b = break_cdf.variables['ninety_five'][:]
ninety_nine_b = break_cdf.variables['ninety_nine'][:]
z_b = break_cdf.variables['z'][:]
break_cdf.close()

# Monsoon
monsoon_cdf = Dataset(monsoon_file, 'r')
mean_m = monsoon_cdf.variables['mean'][:]
median_m = monsoon_cdf.variables['median'][:]
ninety_m = monsoon_cdf.variables['ninety'][:]
ninety_five_m = monsoon_cdf.variables['ninety_five'][:]
ninety_nine_m = monsoon_cdf.variables['ninety_nine'][:]
z_m = monsoon_cdf.variables['z'][:]
break_cdf.close()

# Break - seaparted by organization
break_cdf = Dataset(break_file_sea, 'r')
mean_b_organized = break_cdf.variables['mean'][:]
median_b_organized = break_cdf.variables['median'][:]
ninety_b_organized = break_cdf.variables['ninety'][:]
ninety_five_b_organized = break_cdf.variables['ninety_five'][:]
ninety_nine_b_organized = break_cdf.variables['ninety_nine'][:]

mean_b_disorganized = break_cdf.variables['mean'][:]
median_b_disorganized = break_cdf.variables['median'][:]
ninety_b_disorganized = break_cdf.variables['ninety'][:]
ninety_five_b_disorganized = break_cdf.variables['ninety_five'][:]
ninety_nine_b_disorganized = break_cdf.variables['ninety_nine'][:]
z_b = break_cdf.variables['z'][:]
break_cdf.close()

# Monsoon - seaparted by organization
monsoon_cdf = Dataset(monsoon_file_sea, 'r')
mean_m_organized = monsoon_cdf.variables['mean'][:]
median_m_organized = monsoon_cdf.variables['median'][:]
ninety_m_organized = monsoon_cdf.variables['ninety'][:]
ninety_five_m_organized = monsoon_cdf.variables['ninety_five'][:]
ninety_nine_m_organized = monsoon_cdf.variables['ninety_nine'][:]

mean_m_disorganized = monsoon_cdf.variables['mean'][:]
median_m_disorganized = monsoon_cdf.variables['median'][:]
ninety_m_disorganized = monsoon_cdf.variables['ninety'][:]
ninety_five_m_disorganized = monsoon_cdf.variables['ninety_five'][:]
ninety_nine_m_disorganized = monsoon_cdf.variables['ninety_nine'][:]
z_b = monsoon_cdf.variables['z'][:]
monsoon_cdf.close()
```

## W pdfs by Drosdowsky not separated by organization

```python
plt.figure(figsize=(14,6))
plt.subplot(121)
plt.plot(mean_m,z_m,linewidth=2)
plt.plot(median_m,z_m,linewidth=2)
plt.plot(ninety_m,z_m,linewidth=2)
plt.plot(ninety_five_m, z_m,linewidth=2)
plt.plot(ninety_nine_m, z_m,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Monsoon')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')
plt.legend(['Mean', 'Median', '90', '95', '99'])

plt.subplot(122)
plt.plot(mean_b,z_m, linewidth=2)
plt.plot(median_b,z_m,linewidth=2)
plt.plot(ninety_b,z_m, linewidth=2)
plt.plot(ninety_five_b, z_m,linewidth=2)
plt.plot(ninety_nine_b, z_m,linewidth=2)
plt.xlim([0,30])
plt.ylim([0,20000])
plt.title('Break')
plt.xlabel('W [m/s]')
plt.ylabel('Z [m]')
```

```python
plt.figure(figsize=(14,14))
plt.subplot(321)
plt.plot(median_b,z_b,linewidth=2)
plt.plot(median_m,z_m,linewidth=2)
plt.legend(['Break', 'Monsoon'], loc='best')
plt.xlim([0,1e10])
plt.title('Median')
plt.xlabel('Vertical mass flux [kg/s]')
plt.ylabel('Z [m]')

plt.subplot(322)
plt.plot(mean_b,z_b,linewidth=2)
plt.plot(mean_m,z_m,linewidth=2)
plt.xlim([0,1e10])
plt.title('Mean')
plt.xlabel('Vertical mass flux [kg/s]')
plt.ylabel('Z [m]')

plt.subplot(323)
plt.plot(ninety_b,z_b,linewidth=2)
plt.plot(ninety_m,z_m,linewidth=2)
plt.xlim([0,1e10])
plt.xlabel('Vertical mass flux [kg/s]')
plt.ylabel('Z [m]')
plt.title('90%')

plt.subplot(324)
plt.plot(ninety_five_b,z_b,linewidth=2)
plt.plot(ninety_five_m,z_m,linewidth=2)
plt.xlim([0,2e10])
plt.xlabel('Vertical mass flux [kg/s]')
plt.ylabel('Z [m]')

plt.title('95%')
plt.subplot(325)
plt.plot(ninety_nine_b,z_b,linewidth=2)
plt.plot(ninety_nine_m,z_m,linewidth=2)
plt.xlim([0,3e10])
plt.title('99%')
plt.xlabel('Vertical mass flux [kg/s]')
plt.ylabel('Z [m]')
```
